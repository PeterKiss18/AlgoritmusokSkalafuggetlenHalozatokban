```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
n = 50
```

```python
seed1 = random.randint(10, 1000)
G = nx.generators.random_graphs.barabasi_albert_graph(n, 1, seed=seed1)
```

```python
d = dict(G.degree)
seed2 = random.randint(10, 1000)
my_pos = nx.spring_layout(G, seed = seed2) 
nx.draw(G, nodelist=d.keys(), pos=my_pos, node_size=[v * 30 for v in d.values()])
plt.show()
```
